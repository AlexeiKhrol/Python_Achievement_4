```python
df = ords_prods_merge[:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['prices'].max()
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15 & (df['prices'] > 5), 'price_range_loc')] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#2 and #3 Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural).
result_1 = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result_1.append("Busiest days")
  elif value == 1:
    result_1.append("Busiest days")
  elif value == 3:
    result_1.append("Slowest days")
  elif value == 4:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

```python
ords_prods_merge['busiest_days'] = result_1
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
#4 When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day']==0) & (ords_prods_merge['order_hour_of_day']>=1) & (ords_prods_merge['order_hour_of_day']<=6), 'busy_time'] = 'Fewest orders'
```
